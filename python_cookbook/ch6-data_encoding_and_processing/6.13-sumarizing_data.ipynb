```python
import pandas as pd

# Read the CSV file
file_path = '/home/tourist/diy/LearnPy/data/rats_synthetic.csv'
rats = pd.read_csv(file_path)

# Inspect the dataset
print(rats.info())
print(rats.head())

# Explore unique values in a column
print(rats['Current Activity'].unique())

# Filter data
crew_dispatched = rats[rats['Current Activity'] == 'Dispatch Crew']
print(f"Number of rows where 'Current Activity' is 'Dispatch Crew': {len(crew_dispatched)}")

# Find top 5 most rat-infested ZIP codes
top_zip_codes = crew_dispatched['ZIP Code'].value_counts().head(5)
print("Top 5 most rat-infested ZIP codes:")
print(top_zip_codes)

# Group by completion date and count entries
dates = crew_dispatched.groupby('Completion Date')
date_counts = dates.size()

# Sort and display the top 10 busiest days
sorted_dates = date_counts.sort_values(ascending=False)
print("Top 10 busiest days:")
print(sorted_dates.head(10))
```
