```python
# countdown.py
import time
import threading
import pickle

class Countdown:
    def __init__(self, n):
        self.n = n
        self.thr = threading.Thread(target=self.run)
        self.thr.daemon = True
        self.thr.start()

    def run(self):
        while self.n > 0:
            print('T-minus', self.n)
            self.n -= 1
            time.sleep(1)

    def __getstate__(self):
        return self.n

    def __setstate__(self, n):
        self.__init__(n)

# Example usage:
c = Countdown(30)

folder_path = "/home/tourist/diy/LearnPy/data/"

# After a few moments
with open(folder_path + 'cstate.p', 'wb') as f:
    pickle.dump(c, f)

# After restarting Python
with open(folder_path + 'cstate.p', 'rb') as f:
    c = pickle.load(f)  # The thread resumes where it left off
```
